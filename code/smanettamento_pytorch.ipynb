```python
# Roba a caso su pytorch
from torch import nn
import torch
import numpy as np
```

```python
model = nn.Sequential(
    nn.Linear(1,1)
)

with torch.no_grad():
    for p in model.parameters():
        print(p)
```

```python
x = torch.tensor([2], dtype=torch.float32)
```

```python
model(x).detach().numpy()
```

```python
model.zero_grad()
```

```python
loss_fun = nn.MSELoss()
loss = loss_fun(torch.tensor([2.0]), model(x)+1)
```

```python
loss.backward()
print(loss)
```

```python
with torch.no_grad():
    for p in model.parameters():
        print(p.grad)
```
